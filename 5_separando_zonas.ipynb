Separando dataframes por zona

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/phcavalcanti/projeto_imobiliaria_rj/main/aluguel_residencial_sem_outliers.csv'
```

```python
dados = pd.read_csv(url, sep = ';')
```

```python
dados.head(10)
```

### zona sul

```python
selecao_sul = dados['Zona'] == 'Sul'#criando selecao somente com bairros que sejam da zona sul com base na lista zona_sul
zona_sul = dados[selecao_sul]#novo df somente com imoveis da zona sul
```

```python
zona_sul
```

```python
zona_sul.to_csv('residencias_zona_sul.csv', sep=';', index = False)
```

### zona norte

```python
selecao_norte = dados['Zona'] == 'Norte'
zona_norte = dados[selecao_norte]
```

```python
zona_norte
```

```python
zona_norte.to_csv('residenciais_zona_norte.csv', sep = ';', index = False)
```

### zona oeste

```python
selecao_oeste = dados['Zona'] == 'Oeste'
zona_oeste = dados[selecao_oeste]
```

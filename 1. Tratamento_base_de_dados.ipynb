#Relatório de análise II

## Selecionando/filtrando os dados para análise

-> Serão selecionados somentes dados correlatos a imóveis residenciais.

```python
tipo_de_imovel = dados['Tipo']
```

```python
tipo_de_imovel
```

```python
tipos_de_imoveis =pd.DataFrame(dados['Tipo'].unique(), columns = ['Tipo de imóvel'])#mostrando somente as categorias do tipo de imóveis
tipos_de_imoveis.columns.name = 'Id'
```

```python
tipos_de_imoveis
```

```python
#criando lista somente com imoveis residenciais
residencial = ['Quitinete',
'Casa',
'Apartamento',
'Casa de Condomínio',
'Casa de Vila']
```

```python
selecao = dados['Tipo'].isin(residencial)
```

```python
dados_residencial = dados[selecao]#criando novo dataframe so com residenciais
```

```python
dados_residencial#DataFrame somente com os imoveis residenciais
```

```python
dados_residencial['Tipo'].unique()#verificando que somente há imoveis residenciais nesse dataframe criado
```

```python
dados_residencial.shape
```

```python
dados_residencial.info()
```

```python
#Reconstruindo index
dados_residencial.index = range(dados_residencial.shape[0])
```

```python
dados_residencial
```

## Exportando base de dados

```python
#exportando base de dados para futuras análises
dados_residencial.to_csv('base_dados_aluguel_residencial.csv', sep = ';', index = False)
```

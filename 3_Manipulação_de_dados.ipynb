```python
def zonas(x):#funcao que vai analisar a coluna bairros e com base nas listas das zonas vai informar qual zona pertence
    if x in zona_sul:
        return "Sul"
    elif x in zona_norte:
        return "Norte"
    elif x in zona_oeste:
        return "Oeste"
    elif x in zona_centro:
        return "Centro"
    else:
        return "Outra Regiao"


```

```python
dados['Zona'] = dados['Bairro'].apply(zonas)
```

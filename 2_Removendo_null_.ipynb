## Tratamento de dados faltantes

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/phcavalcanti/projeto_imobiliaria_rj/main/base_dados_aluguel_residencial.csv'
```

```python
dados = pd.read_csv(url, sep = ';')
```

```python
dados.head(10)
```

```python
pd.DataFrame(dados.info())
```

Pela Tabela acima, nota se a presença de valores nulos para aluguel(Valor), condominio e IPTU, logo deve-se retirar valores nulo para aluguel, visto que não terá utilidade para a análise à doravante.

```python
dados[dados['Valor'].isnull()]#dataframe com valores nulos para o aluguel
```

```python
dados.dropna(subset =['Valor'], inplace = True)#removendo valores nulos
```

```python
dados[dados['Valor'].isnull()]#nao ha mais dados nulos para valor
```

## tratando os nulos para condomínio e iptu

1. Se for apartamento e tiver condomínio nulo, será excluído.

2. Se for demais tipos de imovel, atribuir zero ao valor do condomínio

```python
dados[dados['Condominio'].isnull()].shape[0]#quantas entradas com valores nulos de condomínio
```

```python
#selecionando apartamentos com condomínio nulo.
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())# o que deve ser excluído
```

```python
dados[selecao]
```

```python
dados = dados[~selecao]#selecionando somente o que esta fora da selecao, isto é, o que não é apartamento com condominio nulo, através da negacao ~.
```

```python
dados
```

```python
dados[selecao]#nao ha mais apartamentos sem condominio
```

```python
dados[dados['Condominio'].isnull()].shape[0]#demais tipos de imoveis com condominio igual a zero
```

```python
dados.fillna({'Condominio':0, 'IPTU':0}, inplace = True)#substituindo onde ha null por zero
```

```python
dados.info()#nao ha mais valores nulos
```

```python
dados.to_csv('alugueis_residenciais.csv', sep = ';', index = False)#exportando dados tratados
```
